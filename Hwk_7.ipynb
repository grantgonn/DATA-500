### Exercise 1(a) (4 points)

Create a function called `calculate_total_sales(sales_list)` that:

- Takes a list of numbers representing sales transactions.
- Returns the total sales amount (sum of all values in `sales_list`).

```python
import pandas as pd
import numpy as np
```

```python
def calculate_total_sales(sales_list):
    return np.sum(sales_list)
```

### Exercise 1(b) (6 points)

Create a function called `test_calculate_total_sales(sales_list)` that:

- Test if the input is a list.
- Test with an empty list (should return `0`).
- Test with a list containing one transaction.
- Test with a list of multiple transactions.

```python
def test_fun():

    #test 1
    assert isinstance([10,30,25,15], list), 'input not a list'
    #assert isinstance(4, list), 'input not a list'
        
    #test 2
            
    if (calculate_total_sales([]), 0):
        print('output is 0')
        
    #test 3
    print(calculate_total_sales([5]))
    
    #test 4
    print(calculate_total_sales([5,13,20]))
    
    print('all test passed')
    
test_fun()
```

### Exercise 1(c) (4 points)

Create a function called `calculate_average_sales(sales_list)` function that:

- Takes a list of sales transactions.
- Returns the average sales amount. If `sales_list` is empty, return `0`.

```python
def calculate_average_sales(sales_list):
    if not sales_list:
        return 0
    else:
        return np.mean(sales_list)
```

### Exercise 1(d) (6 points)

Create a function called `test_calculate_average_sales(sales_list)` that:

- Test if the input is a list.
- Test with an empty list (should return `0`).
- Test with a list containing one transaction.
- Test with a list of multiple transactions.

```python
def test_fun():

    #test 1
    assert isinstance([10,30,25,15], list), 'input not a list'
    #assert isinstance(4, list), 'input not a list'
        
    #test 2
            
    if (calculate_average_sales([]), 0):
        print('output is 0')
        
    #test 3
    print(calculate_average_sales([5]))
    
    #test 4
    print(calculate_average_sales([5,13,20]))
    
    print('all test passed')
    
test_fun()
```

### Exercise 1(e) (6 points)

Create a function called `identify_outliers(sales_list, threshold=1.5)` that:

- Identifies any transaction that is above `threshold * mean` as an outlier.
- Returns a list of outliers.

Notice that you need to use `calculate_average_sales` from part 1(c) to compute the average.

```python
def identify_outliers(sales_list, threshold=1.5):
    outliers = []
    mean_value = np.mean(sales_list)
    for sale in sales_list:
        if sale > threshold * mean_value:
            outliers.append(sale)
    return outliers
```

### Exercise 1(f) (6 points)

Create a function called `test_identify_outliers(sales_list, threshold=1.5)` that:

- Test with no outliers.
- Test with one or more outliers.
- Test with an empty list.

```python
def test_fun():

    #test 1
    print(identify_outliers([1,2,3,4,5]))
    
    #test 2
    print(identify_outliers([1,2,3,4,5,100]))
    
    #test 3
    print(identify_outliers([]))
    
    print('all test passed')
    
test_fun()
```

### Exercise 2(b) (2 points)

```python
df = pd.read_csv('powerconsumption.csv')
df.head()
```

### Exercise 2(c) (3 points)

Convert `Datetime` to date-time format. 

```python
df['Datetime'] = pd.to_datetime(df['Datetime'])
df.head()
```

### Exercise 2(d) (8 points)

Create a time-series plot that shows how `Temperature` and `Humidity` evolve with time.

```python
fig, ax = plt.subplots(figsize = (12,8))

sns.lineplot(x='Datetime', y='Temperature', data=df, ax=ax)
sns.lineplot(x='Datetime', y='Humidity', data=df, ax=ax)
plt.title('Tempurature and Humidity over time')
plt.xlabel('Date')
plt.ylabel('Temperature and Humidity')
plt.show()
```

### Exercise 2(e) (6 points)

Create a scatter-plot of `Temperature` and `Humidity`. Describe the chart.

```python
fig, ax = plt.subplots(figsize = (12,8))

sns.scatterplot(x='Humidity', y='Temperature', data=df, ax=ax)
plt.title('Tempurature and Humidity')
plt.xlabel('Humidity')
plt.ylabel('Temperature')
plt.show()
```

From the above chart we can see a negative correlation. As Temp decreases humidity increases

### Exercise 2(f) (5 points)

Create two columns in the `df` data-frame called:

- `Date`: extract the date (wihout time) from `Datetime` and store it in this column.
- `Day`: extract the day (three letter abbrevation) from `Datetime` and store it in this column.

```python
df['Date'] = df['Datetime'].dt.date
df['Day'] = df['Datetime'].dt.day
df.head()
```

### Exercise 2(g) (8 points)

Create a data-frame called `df_date_mean`. This data-frame should contains the average `Temperature` and average `Humidity` based on `Date`. Then, create a scatter plot of `Temperature` and `Humidity`. Describe the plot.

```python
df_date_mean = df.groupby('Date').agg({'Temperature': 'mean', 'Humidity': 'mean'})

fig, ax = plt.subplots(figsize = (12,8))

sns.scatterplot(x='Humidity', y='Temperature', data=df_date_mean, ax=ax)
plt.title('Tempurature and Humidity')
plt.xlabel('Humidity')
plt.ylabel('Temperature')
plt.show()
```

There is a slight negative correlation but for the most part there is no trend
