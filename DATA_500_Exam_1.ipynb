# Exercise 5 (6 points)

Consider the data frame presented in Exercise 3. Write a function that takes the data frame from Exercise 3, and returns the mean and standard deviation of the numeric columns.

````python
import pandas as pd

data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'Salary': [50000, 60000, 70000]
}

data = pd.DataFrame(data)

def stats(df):
    mean_age = df['Age'].mean()
    mean_salary = df['Salary'].mean()
    std_age = df['Age'].std()
    std_salary = df['Salary'].std()
    return mean_age, mean_salary, std_age, std_salary

mean_age, mean_salary, std_age, std_salary = stats(data)

print('mean age:', mean_age)
print('mean salary:', mean_salary)
print('std age:', std_age)
print('std salary:', std_salary)
````

````python
import pandas as pd

# Create a sample DataFrame
data = {'Name': ['Alice', 'Bob', 'Charlie'],
        'Age': [25, 30, 35],
        'Salary': [50000, 60000, 70000]}

df = pd.DataFrame(data)

# Define a function to increase salary by 10%
def increase_salary(salary):
    return salary * 1.1

# Apply the function to the Salary
df['New_Salary'] = df['Salary'].apply(increase_salary)

# Print the updated DataFrame
print(df)
````
