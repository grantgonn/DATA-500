### Exercise 1(b) (2 points)

Read the `diamonds` data file, and create a data frame called `diamonds`.

### Exercise 1(c) (2 points)

Create a lambda function called `carat_flag` that flags diamonds with `carat` values greater than 2.

```python
carat_flag = lambda x: x > 2
```

```python
diamonds['flagging_carat'] = diamonds['carat'].apply(carat_flag)
diamonds[diamonds['flagging_carat'] == True].head()
```

### Exercise 1(e) (6 points)

Convert the `cut`, `color`, and `clarity` columns to 0-1 columns, and append them to the `diamonds` data frame. Then, drop `cut`, `color`, and `clarity` from the data frame.

```python
dummies = pd.get_dummies(diamonds[['cut','color','clarity']], dtype=int, drop_first=True)
dummies.head()
```

```python
diamonds = pd.concat([diamonds, dummies], axis=1)
diamonds.head()
```

```python
diamonds = diamonds.drop(columns=['cut', 'color', 'clarity'], axis=1)
diamonds.head()
```

### Exercise 1(f) (4 points)

Create a lambda function called  `ratio` that computes the ratio of `depth` and `table`. Then, create a column called `depth/table`, using the `apply` function, that stores the values of `ratio` applied to the `diamonds` data frame.

```python
ratio = lambda row: row['depth'] / row['table']

diamonds['depth/table'] = diamonds.apply(ratio, axis=1)

diamonds.head()
```

### Exercise 1(g) (4 points)

Create a lambda function called `vol` that computes the product of `x`, `y`, and `z`. Then, create a column called `volume`, using the `apply` function, that stores the values of `vol` applued to the `diamonds` data frame.

```python
vol = lambda row: row['x'] * row['y'] * row['z']

diamonds['volume'] = diamonds.apply(vol, axis=1)

diamonds.head()
```

### Exercise 2(a) (12 points)

Create a class called `DataProcessing`. The class should include the following methods:

- `__init__(self, data: pd.DataFrame)`: Initializes the class with a pandas DataFrame.
- `encode_categorical(self, columns)`: Encodes categorical columns using `get_dummies` encoding from `pandas`. 
- `volume(self)`: Computes the volume of the diamond as `x * y * z`.
- `ratio(self)`: Computes the ratio of `depth` and `table`.
- `stand(self)`: Computes the z-score standardization of `carat`. That is, `(carat - mean(carat)) / std(carat)`, where `mean` and `std` are the mean and standard deviation, respectively.
- `fe(self)`: Append the `volume` and `ratio` engineered features to the data frame. Also apply `stand`.

Implement the `DataProcessing` class with the above methods and demonstrate its usage with a sample DataFrame.


```python
class DataProcessing:
    def __init__(self, data: pd.DataFrame):
        self.data = data
        
    def encode_categorical(self, columns):
        dummies = pd.get_dummies(self.data[[columns]], dtype=int, drop_first=True)
        self.data = pd.concat([self.data, dummies], axis=1)
        self.data = self.data.drop(columns=[columns], axis=1)
        return self
        
    def volume(self):
        self.data['volume'] = self.data['x'] * self.data['y'] * self.data['z']
    
    def ratio(self):
        self.data['depth/table'] = self.data['depth'] / self.data['table']
    
    def stand(self):
        self.data['stand'] = (self.data['carat'] - self.data['carat'].mean()) / self.data['carat'].std()
    
    def fe(self):
        self.volume()
        self.ratio()
        self.stand()
    
```

### Exercise 2(c) (3 points)

Initialize the `DataProcessing` class, and then run the `fe` method on the data frame. 

```python
diamonds_processed = DataProcessing(diamonds)

diamonds_processed.fe()
diamonds_processed.data.head()
```
