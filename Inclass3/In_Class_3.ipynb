Load `pandas` and `numpy`

# Question 2 (2 points)

Create a data-frame called `chipotle` using the data from the `chipotle.csv`.

```python
chipotle = pd.read_csv('chipotle.csv')
chipotle.head()
```

# Question 3 (8 points)

How many products cost more than $10.00?

```python
chipotle['item_price'] = chipotle['item_price'].astype(str)
chipotle['item_price'] = chipotle['item_price'].str.replace('$','').astype(float)
chipotle.head()
```

```python
chipotle_filter = chipotle.drop_duplicates(['item_name', 'quantity', 'choice_description'])
chipotle_filter.head()
```

```python
chipotle_one = chipotle_filter[chipotle_filter['quantity']==1]
chipotle_one[chipotle_one['item_price']>10]
```

```python
chipotle_one[chipotle_one['item_price']>10]['item_name'].unique()
```

# Question 4 (2 points)

Sort by the name of the item

```python
chipotle.sort_values(by='item_name')
```

# Question 5 (2 points)

What was the quantity of the most expensive item ordered?

```python
chipotle.sort_values(by='item_price', ascending=False)
```

# Question 6 (2 points)

How many times was a Veggie Salad Bowl ordered?

```python
chipotle[chipotle['item_name']=='Veggie Salad Bowl'].shape
```

# Question 7 (3 points)

How many times did someone order more than one Canned Soda?

```python
chipotle[(chipotle['item_name']=='Canned Soda') & (chipotle['quantity']> 1)].shape
```
