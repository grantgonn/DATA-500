### Exercise 2(d) (5 points)

Does `Hours_Studied` have an effect on `Exam_Score`? Create a scatter-plot to explore the relationship between `Hours_Studied` and `Exam_Score`. 

````python
fig, ax = plt.subplots(figsize = (10,6))

ax.scatter(student['Hours_Studied'], student['Exam_Score'])
ax.set_title('Hours Studied & Exam Score')
ax.set_xlabel('Hours Studied')
ax.set_ylabel('Exam Score')
````

From the above chart we can see a general positive correlation betwen exam score and hours studied

### Exercise 2(e) (5 points)

Does `Parental_Involment` play a role on `Exam_Score`? Create a side-by-side boxplot that compares the `Exam_Score` of the different `Parental_Involvement` labels. 

````python
student['Parental_Involvement'].value_counts()
````

````python
inv = ['Low', 'Medium', 'High']
data = [student[student['Parental_Involvement']==iv]['Exam_Score']for iv in inv]

fig, ax = plt.subplots(figsize = (10,6))

ax.boxplot(data, labels=inv, patch_artist=True)
ax.set_title('Exam Score by Parental Involvemnt')
ax.set_xlabel('Parental Involvement')
ax.set_ylabel('Exam Score')
plt.show();
````

From the above chart we can see a slight increase in exam scores as parental involvemnt increases.

### Exercise 2(g) (10 points)

Is there a realtionship between `Parental_Involvement`, `Teacher_Quality`, and `Exam_Score`? Create a heatmap that shows the average `Exam_Score` based on `Parental_Involvement` and `Teacher_Quality`. Describe the chart.

````python
import seaborn as sns

fig, ax = plt.subplots(figsize = (12,8))

#define catagorical features
cat_feature_1 = 'Parental_Involvement'
cat_feature_2 = 'Teacher_Quality'
score = 'Exam_Score'

pivot = student.pivot_table(index=cat_feature_1, columns=cat_feature_2, values=score, aggfunc='mean')

sns.heatmap(pivot, annot=True, ax=ax, cmap='coolwarm')
plt.title('parental involvement and teacher quality by exam score')
plt.xlabel('parental involvement')
plt.ylabel('teacher quality')
plt.show()
````

From the above chart we can see see the realationships are all pretty similar from 66-69 depending on parental involvemt and teacher quality. where the lowest scores are with both low quality and involvment and the highs with high quality and involvment.
